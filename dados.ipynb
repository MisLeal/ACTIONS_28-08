```python
import pandas as pd
```

```python
dados_capacity = pd.read_csv('capacity.csv')
dados_crm = pd.read_csv('resultado_consulta.csv', sep='|')
```

```python
dados_capacity.head()
```

```python
dados_crm.head()
```

```python

dados_crm.rename(columns={'Operador': 'Nome'}, inplace=True)
```

```python
##Dados Do CRM prontos
crm_quantificado = dados_crm.groupby('Nome')['Codigo_do_Historico'].count().reset_index(name='Qtd_Historicos')
crm_quantificado
```

```python
df1 = pd.read_csv('2025-08-28_LEAL_ACTIONS_CONSUMER.csv', sep='|')
df2 = pd.read_csv('2025-08-28_LEAL_ACTIONS_DE.csv', sep='|')
df3 = pd.read_csv('2025-08-28_LEAL_ACTIONS_MERCHANT.csv', sep='|')

# Combinação e pré-processamento
df_combined = pd.concat([df1, df2, df3], ignore_index=True)
df_combined = df_combined[['cust_id', 'agente_accion', 'tipo_contacto', 'fecha_accion']]
df_combined['fecha_accion'] = pd.to_datetime(df_combined['fecha_accion'])

# Agrupamento principal
df_grouped = df_combined.groupby(['tipo_contacto', 'agente_accion']).size().reset_index(name='Total')
```

```python
df_grouped.head()
```

```python
df_combined.rename(columns={'agente_accion': 'Usuario'}, inplace= True)
```

```python
dados_capacity.head()
```

```python
dados_capacity.rename(columns={'USUÁRIO ENGAGE':'Usuario'})
```

```python
df_grouped.rename(columns={'agente_accion':'Usuario'})
df_grouped.info()
```

```python
dados_capacity.head()
```

```python
dados_analisados = pd.merge(df_grouped, dados_capacity, left_on = 'agente_accion', right_on='USUÁRIO ENGAGE', how='inner')
```

```python
dados_analisados.head()
```

```python
dados_analisados = dados_analisados[['tipo_contacto', 'agente_accion', 'NOME', 'Total']]
```

```python
dados_analisados.to_csv('dados_analisados.csv', index=False)
```
